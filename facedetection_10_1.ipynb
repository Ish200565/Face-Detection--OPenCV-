```python
#face DEtection
#import OPencv library for image processing
import cv2
#import cv_2imshow to display images in google collab
from google.colab.patches import cv2_imshow
#import numpy to work with arrays
import numpy as np
```

```python
#Step 1 -> Load the input image

#Path of the image file(update this path as needed)
img_path='group.png'

#Read the image from the given path
#OpenCV loads images in BGR forest
img=cv2.imread(img_path)

#Check if the image was loaded successfully or not
#If the path is wrong or file not found,img will be None
if img is None:
  print(f'Error:Unable to load image from {img_path}')
else:
  print(f'Image loaded successfully from {img_path}')

exit()


```

```python
import cv2 # Re-import cv2 as it might have been cleared by a kernel restart

#Step 2>>Load Haar Cascade for face detection
#Load the pretrained Haar CAscadeXML file for frontal face detection

#cv2.data.haarcascades gives the default path where OpenCV stores these files
face_cascade=cv2.CascadeClassifier(
    cv2.data.haarcascades + 'haarcascade_frontalface_default.xml' # Corrected typo: harrasCascades to haarcascades
)
```

```python
#Step 3 COnvert image to grayscale
#haar Cascade works on grayscale images,not color images

# Re-define 'img' variable as it was cleared by the 'exit()' call in a previous cell.
# This ensures 'img' is available for cv2.cvtColor().
# The original image path is assumed to be 'image.png' as per cell Djkk9pECO_j0.
img_path = 'group.png'
img = cv2.imread(img_path)

# Proceed with grayscale conversion only if the image was loaded successfully
if img is not None:
  gray_img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
else:
  print(f"Error: Could not load image from {img_path}. Grayscale conversion skipped.")
```

```python
#step 4--> Detcet faces in the image
#detectMultiscale() dtects ojects(faces) of diff sizes

faces=face_cascade.detectMultiScale(
    gray_img, #Grayscale image
    scaleFactor=1.1, #HOW MUCH THEIMAGE SIZE IS REDUCED AT RECAH SCALE
    minNeighbors=5 ,  #Mini mum size is detctd face
)
```

```python
#Step 5 DRaw rectangles around faces
#Loop through all detected faces
for(x,y,w,h) in faces:
  #draw rectangles around each detected face (255,0,0)-> blue color in bgr format 2
 #32->thickness of rectangle
  cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
```

```python
#step 6
from google.colab.patches import cv2_imshow # Re-import cv2_imshow
cv2_imshow(img) # Display the image variable, not the filename string
```

```python
#step 7 save the image output
cv2.imwrite('detected_faces.jpg',img)
```
